## Lab 3 - Bài 1

```python
# Code lab 3 bài 1 ở đây
# Viết chương trình thực hiện các thao tác sau với một dãy số nguyên. 
# ✓ Sắp xếp theo thứ tự tăng dần và xuất dãy số nguyên ra màn hình. 
# ✓ Xuất phần tử có giá trị nhỏ nhất ra màn hình 
# ✓ Tính và xuất ra màn hình trung bình cộng các phần tử chia hết cho 3 

day_so=[]
while True:
    so=input("Nhập số: ")
    if so.lower() == "q":
        print("Thoát chương trình")
        break
    else :
        try:
            phan_tu=int(so)
            day_so.append(phan_tu)
        except ValueError:
            print("Không phải số, vui lòng nhập lại!")
print(day_so)

day_so.sort()

print("Dãy số tăng dần:", day_so)
print("Phần tử có giá trị nhỏ nhất:", day_so[0])
print("Phần tử có giá trị lớn nhất:", day_so[-1])


tong=0
so_phan_tu=0
for gia_tri in day_so:
    if gia_tri % 3 == 0:
        tong += gia_tri
        so_phan_tu += 1
trung_binh_cong = tong / so_phan_tu
print(f"Trung bình cộng của các số chia hết cho 3: {trung_binh_cong}")

# Thay đổi sắp xếp theo thứ tự giảm dần, khi đó phần tử nhỏ nhất trong danh 
# sách sẽ là phần tử thứ mấy? 
day_so.reverse()
min = min(day_so)
vi_tri = day_so.index(min)
print( f"vị trí số bé nhất trong dãy: {vi_tri}")
```

## Lab 3 - Bài 5: bài tập làm thêm

```python
#Xây dựng chương trình máy tính đơn giản với các chức năng sau:
# •  Phép tính cơ bản: cộng (+), trừ (–), nhân (×), chia (÷).
# •  Lũy thừa: tính xyx^yxy.
# •  Căn bậc hai: tính x\sqrt{x}x.
# •  Hàm lượng giác: sin, cos, tan.
# •  Logarit:
# •	 Log cơ số 10 (log10)
# •	 Tự nhiên (ln)
# •	 Log cơ số tùy chọn (log_a(x)).
# •  Giải phương trình bậc nhất: ax+b=0ax + b = 0ax+b=0.
# •  Giải phương trình bậc hai: ax2+bx+c=0ax^2 + bx + c = 0ax2+bx+c=0.
# •  Lịch sử: lưu và xem lại các phép tính đã thực hiện.
# •  Thời gian: hiển thị thời gian hiện tại.
# •  Giao diện thân thiện: menu tiếng Việt, dễ dùng.
import math
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~MENU~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("| 1. Phép tính cơ bản: cộng (+), trừ (–), nhân (×), chia (÷).|")
print("| 2. Lũy thừa: tính x^y.                                     |")
print("| 3. Căn bậc hai: tính √x.                                   |")
print("| 4. Hàm lượng giác: sin, cos, tan.                          |")
print("| 5. Logarit: log10, ln, log_a(x).                           |")
print("| 6. Giải phương trình bậc nhất: ax + b = 0.=                |")
print("| 7. Giải phương trình bậc hai: ax^2 + bx + c = 0.           |")
print("| 8. Lịch sử: xem lại các phép tính đã thực hiện.            |")
print("| 9. Thời gian: hiển thị thời gian hiện tại.                 |")
print("| 0. Thóa chương trình.                                      |")
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
lich_su=[] #lưu lịch sử các phép tính theo disctionary ("STT": , "Lựa chọn": ,"Phép tính":, "Kết quả:"" )

while True:
    try:
        lua_chon=int(input("Mời bạn chọn chức năng (1-9) hoặc 0 để thoát: "))
        if lua_chon== 1 :
            print("Bạn đã chọn phép tính cơ bản")
            phep_tinh=input("Nhập phép tính (+, -, *, /): ")
            a=float(input("Nhập số thứ nhất: "))
            b=float(input("Nhập số thứ hai: "))
            if phep_tinh=="+":
                ket_qua=a+b
                print(f"Kết quả: {a} + {b} = {ket_qua}")
            elif phep_tinh=="-":
                ket_qua=a-b
                print(f"Kết quả: {a} - {b} = {ket_qua}")
            elif phep_tinh=="*":
                ket_qua=a*b
                print(f"Kết quả: {a} * {b} = {ket_qua}")
            elif phep_tinh=="/":
                if b!=0:
                    ket_qua=a/b
                    print(f"Kết quả: {a} / {b} = {ket_qua}")
                else:
                    print("Lỗi: Không thể chia cho 0.")
                    continue
            else:
                print("Phép tính không hợp lệ.")
                continue
            lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"{a} {phep_tinh} {b}", "Kết quả": ket_qua})
        elif lua_chon==2:
            print("Bạn đã chọn lũy thừa")
            x=float(input("Nhập cơ số x: "))
            y=float(input("Nhập số mũ y: "))
            ket_qua= x**y
            print(f"Kết quả: {x}^{y} = {ket_qua}")
            lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"{x}^{y}", "Kết quả": ket_qua})
        elif lua_chon==3:   
            print("Bạn đã chọn căn bậc hai")
            x=float(input("Nhập số x: "))
            if x>=0:
                ket_qua=math.sqrt(x)
                print(f"Kết quả: √{x} = {ket_qua}")
                lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"√{x}", "Kết quả": ket_qua})
            else:
                print("Lỗi: Không thể tính căn bậc hai của số âm.")
                continue
        elif lua_chon==4:       
            print("Bạn đã chọn hàm lượng giác")
            ham_luong_giac=input("Nhập hàm lượng giác (sin, cos, tan): ")
            x=float(input("Nhập góc (đơn vị độ): "))
            radian=math.radians(x) #đổi độ sang radian
            if ham_luong_giac=="sin":
                ket_qua=math.sin(radian)
                print(f"Kết quả: sin({x}) = {ket_qua}")
            elif ham_luong_giac=="cos":
                ket_qua=math.cos(radian)
                print(f"Kết quả: cos({x}) = {ket_qua}")
            elif ham_luong_giac=="tan":
                ket_qua=math.tan(radian)
                print(f"Kết quả: tan({x}) = {ket_qua}")
            else:
                print("Hàm lượng giác không hợp lệ.")
                continue
            lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"{ham_luong_giac}({x})", "Kết quả": ket_qua})
        elif lua_chon==5:       
            print("Bạn đã chọn logarit")
            loai_log=input("Nhập loại logarit (log10, ln, log_a): ")
            if loai_log=="log10":
                x=float(input("Nhập số x (x > 0): "))
                if x>0:
                    ket_qua=math.log10(x)
                    print(f"Kết quả: log10({x}) = {ket_qua}")
                    lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"log10({x})", "Kết quả": ket_qua})
                else:
                    print("Lỗi: x phải lớn hơn 0.")
                    continue
            elif loai_log=="ln":
                x=float(input("Nhập số x (x > 0): "))
                if x>0:
                    ket_qua=math.log(x)
                    print(f"Kết quả: ln({x}) = {ket_qua}")
                    lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"ln({x})", "Kết quả": ket_qua})
                else:
                    print("Lỗi: x phải lớn hơn 0.")
                    continue
            elif loai_log=="log_a":
                x=float(input("Nhập số x (x > 0): "))
                a=float(input("Nhập cơ số a (a > 0 và a != 1): "))
                if x>0 and a>0 and a!=1:
                    ket_qua=math.log(x,a)
                    print(f"Kết quả: log_{a}({x}) = {ket_qua}")
                    lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"log_{a}({x})", "Kết quả": ket_qua})
                else:
                    print("Lỗi: x phải lớn hơn 0, a phải lớn hơn 0 và khác 1.")
                    continue
            else:
                print("Loại logarit không hợp lệ.")
                continue
        elif lua_chon==6:       
            print("Bạn đã chọn giải phương trình bậc nhất")
            a=float(input("Nhập hệ số a (a != 0): "))
            b=float(input("Nhập hệ số b: "))
            if a!=0:
                ket_qua=-b/a
                print(f"Kết quả: Phương trình có nghiệm x = {ket_qua}")
                lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"{a}x + {b} = 0", "Kết quả": ket_qua})
            else:
                print("Lỗi: a không được bằng 0.")
                continue
        elif lua_chon==7:       
            print("Bạn đã chọn giải phương trình bậc hai")
            a=float(input("Nhập hệ số a (a != 0): "))
            b=float(input("Nhập hệ số b: "))
            c=float(input("Nhập hệ số c: "))
            if a!=0:
                delta=b**2-4*a*c
                if delta>0:
                    x1=(-b+math.sqrt(delta))/(2*a)
                    x2=(-b-math.sqrt(delta))/(2*a)
                    print(f"Kết quả: Phương trình có hai nghiệm phân biệt x1 = {x1}, x2 = {x2}")
                    lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"{a}x^2 + {b}x + {c} = 0", "Kết quả": f"x1 = {x1}, x2 = {x2}"})
                elif delta==0:
                    x=-b/(2*a)
                    print(f"Kết quả: Phương trình có nghiệm kép x = {x}")
                    lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"{a}x^2 + {b}x + {c} = 0", "Kết quả": f"x = {x}"})
                else:
                    print("Kết quả: Phương trình vô nghiệm.")
                    lich_su.append({"STT": len(lich_su)+1, "Lựa chọn": lua_chon, "Phép tính": f"{a}x^2 + {b}x + {c} = 0", "Kết quả": "Vô nghiệm"})
            else:
                print("Lỗi: a không được bằng 0.")
                continue
        elif lua_chon==8:       
            print("Lịch sử các phép tính đã thực hiện:")
            if len(lich_su)==0:
                print("Chưa có phép tính nào được thực hiện.")
            else:
                for item in lich_su:
                    print(f"{item['STT']}. Lựa chọn: {item['Lựa chọn']}, Phép tính: {item['Phép tính']}, Kết quả: {item['Kết quả']}")
        elif lua_chon==9:       
            import datetime
            now = datetime.datetime.now()
            print("Thời gian hiện tại:", now.strftime("%Y-%m-%d %H:%M:%S"))
        elif lua_chon==0:
            print("Hẹn gặp lại.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
            continue
    except ValueError:
        print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
        continue    
```
