## Lab 6 - Bài 1

```python
# Code lab 6 bài 1 ở đây
from class_lab6 import *
lst_hcn=[]
lst_hv=[]
hinh_chu_nhat=int(input("Nhập số hình chữ nhật muốn tạo: "))
hinh_vuong=int(input("Nhập số hình vuông muốn tạo: "))
for i in range(hinh_chu_nhat):
    print("Nhập hình chữ nhật thứ ",i+1)
    hcn=ChuNhat(int(input(f"Nhập chiều dài hcn {i+1}: ")),int(input(f"Nhập chiều rộng hcn {i+1}: ")))
    lst_hcn.append(hcn)
for i in range(hinh_vuong):
    print("Nhập hình vuông thứ ",i+1)
    canh=int(input(f"Nhập cạnh hình vuông {i+1}: "))
    hv = HinhVuong(canh)
    lst_hv.append(hv)

print("\nHình chữ nhật")
print(f"{"Chiều dài":^10} {"Chiều rộng":^10} {"Chu vi":^10} {"Diện tích":^10}")
for hcn in lst_hcn :
    print(hcn)
    
print("\nHình vuông")
print(f"{"Cạnh":^10} {"Chu vi":^10} {"Diện tích":^10}")
for hv in lst_hv :
    print(hv)
```

## Lab 6 - Bài 2 , 3 , 4 

```python
from class_lab6 import *
danh_sach_sinh_vien= QuanLySinhVien()
menu={
    1 : "Nhập danh sách sinh viên",
    2 : "Xuất thông tin danh sách sinh viên",
    3 : "Xuất danh sách sinh viên học lực giỏi",
    4 : "Sắp xếp danh sách sinh viên theo điểm ",
    5 : "Kết thúc"
}
while True:
    print("~~~~~~~~~~~~~Menu~~~~~~~~~~~~")
    for key,value in menu.items():
        print(f"{key} : {value} ")
    while True:
        try:
            lua_chon= int(input("Nhập lựa chọn"))
            if lua_chon > 0 & lua_chon <= 5 :
                break
            else:
                print("Lựa chọn không hợp lệ")
        except ValueError :
            print("Vui lòng nhập lựa chọn hợp lệ !")
    match lua_chon:
        case 1: 
            danh_sach_sinh_vien.nhap_ds_sinh_vien()
        case 2:
            danh_sach_sinh_vien.xuat_danh_sach()
        case 3:
            danh_sach_sinh_vien.xuat_sv_hoc_luc_gioi()
        case 4:
            danh_sach_sinh_vien.sap_xep_diem()
        case 5:
            print("Kết thúc")
            break
```
