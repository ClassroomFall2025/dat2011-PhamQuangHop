## Lab 6 - Bài 1

```python
# Code lab 6 bài 1 ở đây
from class_lab6 import *
lst_hcn=[]
lst_hv=[]
hinh_chu_nhat=int(input("Nhập số hình chữ nhật muốn tạo: "))
hinh_vuong=int(input("Nhập số hình vuông muốn tạo: "))
for i in range(hinh_chu_nhat):
    print("Nhập hình chữ nhật thứ ",i+1)
    hcn=ChuNhat(int(input(f"Nhập chiều dài hcn {i+1}: ")),int(input(f"Nhập chiều rộng hcn {i+1}: ")))
    lst_hcn.append(hcn)
for i in range(hinh_vuong):
    print("Nhập hình vuông thứ ",i+1)
    canh=int(input(f"Nhập cạnh hình vuông {i+1}: "))
    hv = HinhVuong(canh)
    lst_hv.append(hv)

print("\nHình chữ nhật")
print(f"{"Chiều dài":^10} {"Chiều rộng":^10} {"Chu vi":^10} {"Diện tích":^10}")
for hcn in lst_hcn :
    print(hcn)
    
print("\nHình vuông")
print(f"{"Cạnh":^10} {"Chu vi":^10} {"Diện tích":^10}")
for hv in lst_hv :
    print(hv)
```
